```python
def KNN(X_train,y_train,X_test,y_test):
    test_scores = []
    train_scores = []

    for i in range(2,15):

        knn = KNeighborsClassifier(i)
        knn.fit(X_train,y_train)


        train_scores.append(knn.score(X_train,y_train))
        test_scores.append(knn.score(X_test,y_test))
#     print('Train Scores: ',train_scores)
#     print('Test Scores: ',test_scores)
    return test_scores
```

```python

def ROC_plots(X_train,y_train,X_test,y_test):
    ras=[]
    for i in range(2,15):

        knn = KNeighborsClassifier(i)
        knn.fit(X_train,y_train)

        y_pred =knn.predict_proba(X_test)
        fpr, tpr, threshold = roc_curve(y_test, y_pred[:, 1])


        plt.plot([0,1],[0,1],'k--')
        plt.plot(fpr,tpr, label='Knn')
        plt.xlabel('fpr')
        plt.ylabel('tpr')
        plt.title('Knn ROC curve')
        print(i,'neighbours')
        roc_auc= roc_auc_score(y_test, y_pred[:,1])
#         print(roc_auc)
        ras.append(roc_auc)
        plt.show()
    return ras
```

```python
# # A) Partitioning and K-fold Cross Validation

from sklearn.model_selection import RepeatedKFold 

from sklearn.model_selection import StratifiedKFold

from sklearn.model_selection import KFold 

X = df_arr[:,:8]
y = df_arr[:,-1]

kf = RepeatedKFold(n_splits=6, n_repeats=10, random_state=None) 
X_train,y_train,X_test,y_test=KF(kf)
RKF = KNN(X_train,y_train,X_test,y_test)
print(RKF)
```

```python
## B) Stratified k-fold cross validation:

stkf = StratifiedKFold(n_splits=5, random_state=None)

for train_index, test_index in stkf.split(X,y):
#     print("Train:", train_index, "Validation:",test_index)
#     print(train_index,test_index)
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
# X is the feature set and y is the target
StratKF = KNN(X_train,y_train,X_test,y_test)
print(StratKF)
```
